```python
import pandas as pd 

import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
data=pd.read_csv("Datasheets/Mall_Customers.csv")
data.head()
```

```python
X=data[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
model=KMeans(n_clusters=5)
model.fit(X)
y=model.predict(X)
y
```

```python
Df=X
Df['Class']=y
Df.head()
```

```python
sns.scatterplot(data=Df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Class', palette=['red','green','blue','black','purple'] )
```

```python
model.cluster_centers_
```

```python
error=[]

for i in range(1,10):
    kmodel=KMeans(n_clusters=i)
    kmodel.fit(X)
    y=kmodel.predict(X)
    error.append(kmodel.inertia_)
    
    sns.scatterplot(data=Df, x='Annual Income (k$)', y='Spending Score (1-100)', hue=y,palette=["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd",
    "#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#17becf"] )
    plt.show()
```
