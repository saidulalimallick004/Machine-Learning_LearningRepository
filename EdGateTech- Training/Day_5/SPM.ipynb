```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from Datasheet.svm_margin_plot import plot_svm_boundary

from sklearn.svm import SVC
```

```python
sns.scatterplot(data=data, x='Med_1_mL',y='Med_2_mL' , hue='Virus Present')
```
